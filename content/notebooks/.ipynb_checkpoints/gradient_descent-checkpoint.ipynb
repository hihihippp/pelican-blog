Gradient descent is an algorithm to find the local minimum of a function. Let's suppose we have a simple function, $ f(x) = x^2 âˆ’ 6x + 5$, and we want to find the minimum. We can solve this analytically using calculus, by finding the derivate and setting it to zero:

$$ f'(x) = 2x - 6 $$

$$\begin{align}
2x - 6  = 0\\
2x = 6\\
x = 3\end{align}$$

By simply plotting the function, we can see that the local minimum is indeed at 3:

```python
# <!-- collapse=True -->

x = np.linspace(-15,20,100)
y = x**2-6*x+5

plt.plot(x,y)
plt.plot(3, -4 , "D")
_ = plt.title("Function and local minimum")
```

However, not all functions can be solved for a local minimum as easily as the above function. This is where gradient descent comes in - it's an iterative function using the steepest descent. To find the local minimum, you start at a random point, and move into the direction of steepest descent, i.e. into the direction that goes down. In this example, let's suppose we start at $x = 15$. The gradient at this point is $2 \times 15 - 6 = 24$. Because we're using gradient *descent*, we need to subtract the gradient from our $x$-coordinate. However, notice that $16 - 25$ gives us $-9$, clearly overshooting over target of $-3$. To fix this, we multiply the gradient by a *step size*. This step size (often called gamma) has to be chosen carefully, as a value too small will result in a long computation time, while a value too large will not give you a good result or even fail to converge. In this example, we'll set the step size to 0.01, which means we'll subtract $24 \times 0.01$ from 15, which is $14.76$. This is now our new temporary local minimum: We continue this method until we either don't see a change after we subtracted the gradient * step size, or until we've completed a pre-set number of iterations. 

Below, I've coded this algorithm in Python. The algorithm stops when the values between the new and the temporaroy minimum do not differ by more than 0.001 - if we need more precision, we can decrease this value. According to gradient descent, the local minimum occurs at $3.05$, which is not too far off from the true local minimum.

## Linear regression

Of course, we can also apply gradient descent to slightly more complex problems, such as linear regression. Let's make up some data with noise:

```python
x, y = make_regression(n_samples = 100, n_features=1, n_informative=1, noise=50)
_ = plt.plot(x,y, 'o')
```

Find the line of best fit isn't a problem: We can use least squares to find it analytically:

```python
slope, intercept, _,_,_ = stats.linregress(x[:,0],y)
best_fit = x[:,0] * slope + intercept
plt.plot(x,y, 'o')
_ = plt.plot(x,best_fit, 'k', color = "red")
```

```python
def gradient(x, y, iters, alpha):
    costs = []
    m = y.size # number of data points
    theta = np.ones(2) # 'random' start
    x_trans = x.transpose() # to save computation time
    
    preds = []
    for i in range(iters):
        pred = np.dot(x, theta)
        
        error = pred - y
        cost = np.sum(error ** 2) / (2 * m)
        costs.append(cost)
        
        if i % 25 == 0: preds.append(pred)

        gradient = np.dot(x_trans, error) / m 
        theta = theta - alpha * gradient  # update
    return theta, costs, preds
```

```python
x = np.c_[np.ones(x.shape[0]), x] # insert column necessary for matrix algebra
```

```python
alpha = 0.001 # set step-size
iters = 5000
theta, cost, preds = gradient(x, y, iters, alpha)
```

```python
print("Gradient Descent: {:.2f}, {:.2f}".format(theta[0], theta[1]))
print("Least Squares: {:.2f}, {:.2f}".format(intercept, slope))
```

```python
pred = np.dot(x, theta)
```
